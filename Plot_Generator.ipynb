```python
import matplotlib.pyplot as plt
import numpy as np


generations = [20,31.5,32,33,34,36,38,40]
order_index_values = [
    0.045818948393162945,
    0.045819427219605004,
    0.045818878606052854,
    0.04581830884152778,
    0.045813616269371804,
    0.04581318890042921,
    0.04581033629231059,
    0.045812812319631614
]

# trendline
z = np.polyfit(generations, order_index_values, 1)
p = np.poly1d(z)
# 4. Plot the trend line
# "r--" specifies a red dashed line
plt.plot(generations, p(generations), "r--", label='Trend Line')

plt.plot(generations,order_index_values)
plt.plot(generations,order_index_values,'ko')
plt.xlabel('thousands of generations',size=14)
plt.ylabel('order index',size=14);

# using python to get averages to preserve significant digits
# print((0.04581574496582433+0.0458098796734389)/2)
```
